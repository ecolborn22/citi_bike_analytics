```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# create list holding the months for the csv
year = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
```

```python
# create a list that holds the dataframes for each month
l = []
for x in range(0, 12):
    name = "../data_2018/2018" + year[x] + "-citibike-tripdata.csv"
    l.append(pd.read_csv(name))
    print(len(l[x]))
    cond = (l[x]["birth year"] >= 1948) & (l[x]["gender"] >= 1) & (l[x]["tripduration"] > 180) & (l[x]["tripduration"] < 3600)
    # had to filter the data so that there were 15,000,000 rows or less
    # filter to make sure birth year > 1938 (they would be 70 years old or younger)
    # filter out gender of 0 (unknown)
    # filter out trip duration of over 3 min and under 1 hour
    l[x] = l[x][cond]
    l[x] = l[x].reset_index(drop=True)
    # replace 1 or 2 with M or F respectively
    l[x].loc[(l[x].gender == 1),'gender']='M'
    l[x].loc[(l[x].gender == 2),'gender']='F'
```

```python
# verify that all records from 2018 were extracted properly and filtered
for df in l:
    print(len(df))
```

```python
# preview dataset
l[0].head()
```

```python
# append specific months to create the "QUARTERS"
q = []
count = 0
for y in range(0,4):
    df = l[count]
    for z in range(1,3):
        df = df.append(l[count + z], ignore_index=True)
    q.append(df)
    count += 3
```

```python
# check lengths of each dataframe
print(len(q[0]))
print(len(q[1]))
print(len(q[2]))
print(len(q[3]))
```

```python
# checking on data
q[2].head()
```

```python
# checking on data part 2
q[1].tail()
```

```python
# create entire dataframe for all months
df_2018 = q[0]
for k in range(1,len(q)):
    df_2018 = df_2018.append(q[k], ignore_index=True)
```

```python
# save filtered and compiled data
q[0].to_csv("../data_2018/2018-Q1.csv")
q[1].to_csv("../data_2018/2018-Q2.csv")
q[2].to_csv("../data_2018/2018-Q3.csv")
q[3].to_csv("../data_2018/2018-Q4.csv")
df_2018.to_csv("../data_2018/2018.csv")
```
