```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# create list holding the months for the csv
year = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
```

```python
# create a list that holds the dataframes for each month
l = []
for x in range(0, 12):
    name = "data_2018/2018" + year[x] + "-citibike-tripdata.csv"
    l.append(pd.read_csv(name))
    # filter to make sure birth year > 1900
    # filter out gender of 0 (unknown)
    # filter out trip duration of 60 seconds or less  
```

```python
# verify that all records from 2018 were extracted properly
for df in l:
    print(len(df))
```

```python
# preview dataset
l[0].head()
```

```python
# append specific months to create the "QUARTERS"
q = []
count = 0
for y in range(0,4):
    df = l[count]
    for z in range(1,3):
        df = df.append(l[count + z], ignore_index=True)
    q.append(df)
    count += 3
```
